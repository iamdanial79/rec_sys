```python
import pandas as pd #We import pandas because we want to manuplating the data frame
import numpy as np #Import numpy to perform a wide variety of mathematical operations on arrays
import matplotlib.pyplot as plt #To show the result on data diagram or histogram
from math import sqrt #Because we need it (:
%matplotlib inline
```

```python
# By using pandas library we can store our information int pandas data frame for better use
movies = pd.read_csv(r"C:\Users\aryan computer\Downloads\imbd\movies.csv")
ratings = pd.read_csv(r"C:\Users\aryan computer\Downloads\imbd\ratings.csv")
# Now we can show the first 5 rows by using .head()
movies.head()
```

```python
ratings.head()
```

```python
#So we want to extract the year from title columne and its like movie(year) and because of that we can use some regex
#to extract,d means digits and the formations is like it ignore every things except the year in paranthes
movies['year'] = movies.title.str.extract('(\(\d\d\d\d\))',expand=False) #we add a new columne to dataframe
movies['year'] = movies.title.str.extract('(\d\d\d\d)',expand=False) # removing the paranthese
movies['title'] = movies.title.str.replace('(\(\d\d\d\d\))', '') #removing the year from title
movies['title'] = movies['title'].apply(lambda x: x.strip()) #using strip function to make sure there is no whitespace
movies.head()
```

```python
movies['genres'] = movies.genres.str.split('|') #separating the genres by usin .split methode
movies.head()
```

```python
# here is an intresting things that we want to do
#first we copy our data frame to a new data frame then we do iteration on it 
#for each genres we add columne
#if a movie had that type of genre it gets 1.0 in its row and if it does not it would get a 0.0
moviesWithGenres = movies.copy()
for index, row in movies.iterrows():
    for genre in row['genres']:
        moviesWithGenres.at[index, genre] = 1
moviesWithGenres = moviesWithGenres.fillna(0)
moviesWithGenres.head()
```

```python
#in ratings dataframe we do not need timestamp,so we use .drop methode to get rid of it
ratings = ratings.drop('timestamp',1)
ratings.head()
```

```python
userInput = [
            {'title':'Saving private ryan', 'rating':5},
            {'title':'Soul', 'rating':3.5},
            {'title':'Deadpool', 'rating':3.5},
            {'title':"Inception", 'rating':5},
            {'title':'Parasite', 'rating':4.5}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
#So now we add random movies with the rating 
```

```python
#we wanto to find an id to our new movies by title.
inputId = movies[movies['title'].isin(inputMovies['title'].tolist())]
#By merging it with dataframe we could find oout
inputMovies = pd.merge(inputId, inputMovies)
#Dropping information we won't use from the input dataframe
inputMovies = inputMovies.drop('genres', 1).drop('year', 1)
#Final input dataframe
#If a movie you added in above isn't here, then it might not be in the original 
#dataframe or it might spelled differently, please check capitalisation.
inputMovies
```

```python
#previously we add new movies and again by the diffrence of genres it gets 0 or 1
userMovies = moviesWithGenres[moviesWithGenres['movieId'].isin(inputMovies['movieId'].tolist())]
userMovies
```

```python
#by resetting the index we can avoid future issues
userMovies = userMovies.reset_index(drop=True)
#Because we want to save memory and avoiding new problems 
userGenreTable = userMovies.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
userGenreTable
```

```python
inputMovies['rating']
```

```python
#So now we use .dot to give weights to every genres and see which one is more lilkely to watch
userProfile = userGenreTable.transpose().dot(inputMovies['rating'])
#The user profile
userProfile
```

```python
#Now let's get the genres of every movie in our original dataframe
genreTable = moviesWithGenres.set_index(moviesWithGenres['movieId'])
# droping the unnecessary information
genreTable = genreTable.drop('movieId', 1).drop('title', 1).drop('genres', 1).drop('year', 1)
genreTable.head()
```

```python
#Finding the avarage weights of every movies
recommendationTable = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable.head()
```

```python
#Sort our recommendations in descending order
recommendationTable = recommendationTable.sort_values(ascending=False)
#Lets see it
recommendationTable.head()
```

```python
#The final recommendation table
movies.loc[movies['movieId'].isin(recommendationTable.head(20).keys())]
```
